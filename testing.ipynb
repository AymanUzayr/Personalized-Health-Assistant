```python
import pandas as pd 
import numpy as np 
```

```python
df1 = pd.read_csv("FOOD-DATA-GROUP1.csv")
df2 = pd.read_csv("FOOD-DATA-GROUP2.csv")
df3 = pd.read_csv("FOOD-DATA-GROUP3.csv")
df4 = pd.read_csv("FOOD-DATA-GROUP4.csv")
df5 = pd.read_csv("FOOD-DATA-GROUP5.csv")

df = pd.concat([df1,df2,df3,df4,df5], ignore_index=True)
```

```python
df.head().T
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])

```

```python
df.head()
```

```python
df[['food', 'Caloric Value']].sort_values(by='Caloric Value', ascending=False).head()
```

```python
calorie_cap = df['Caloric Value'].quantile(0.99)
df['Caloric Value'] = np.where(
    df['Caloric Value'] > calorie_cap,
    calorie_cap,
    df['Caloric Value']
)

# Step 2: Replace 0 calorie entries with average of min positive & max (capped) calorie
min_nonzero = df[df['Caloric Value'] > 0]['Caloric Value'].min()
avg_impute_value = (min_nonzero + calorie_cap) / 2

df['Caloric Value'] = df['Caloric Value'].replace(0, avg_impute_value)
```

```python
df[['food', 'Caloric Value']].sort_values(by='Caloric Value', ascending=False).head()
```

```python
import numpy as np

# Define reasonable kcal range for 100g: typically 30 to 600
min_kcal, max_kcal = 30, 600

# Clip calories to the range and rescale macros accordingly
def standardize_to_100g(row):
    calories = row['Caloric Value']
    
    # Clip calories
    clipped_cal = np.clip(calories, min_kcal, max_kcal)
    
    # Scale factor
    scale = clipped_cal / calories if calories > 0 else 1

    # Scale macros
    row['Caloric Value'] = clipped_cal
    row['Protein'] *= scale
    row['Fat'] *= scale
    row['Carbohydrates'] *= scale
    return row

merged_food_df = df.apply(standardize_to_100g, axis=1)
```
